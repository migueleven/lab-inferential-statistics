```python
# Libraries
import math 
import pandas as pd # manipulate dataframes
import numpy as np # numerical python
import matplotlib.pyplot as plt # viz

# New libraries
import scipy.stats as st # stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
# Hypotheses
null_hypothesis = "The mean systolic blood pressure is μ = 120 mm Hg"  # H₀
alt_hypothesis = "The mean systolic blood pressure is not μ = 120 mm Hg"  # H₁

print(f"Null Hypothesis (H₀): {null_hypothesis}")
print(f"Alternate Hypothesis (H₁): {alt_hypothesis}")
```

# Level of significance

```python
# Significance level
alpha = 0.05
print(f"Level of Significance (α): {alpha}")
```

# Calculate Test Statistic

```python
# Calculate the test statistic (z-score)
sample_mean = 130.1
pop_mean = 120
n = 100
pop_std = 21.21

z_stat = (sample_mean - pop_mean) / (pop_std / (n**0.5))
print(f"Test Statistic (z): {z_stat:.2f}")
```

# Calculte the P-value

```python
p_value = 2 * (1 - st.norm.cdf(abs(z_stat)))  # two tailed test since we want to know if there is diferent, not lower
print(f"P-Value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the Null Hypothesis: Significant result.")
else:
    print("Fail to Reject the Null Hypothesis: Not a significant result.")
```

# Critical Value

```python
# Critical Value for a one-tailed test
# Calcular valores críticos para dos colas
z_critical = st.norm.ppf(1 - alpha / 2)  # Critical value positivo
z_critical_neg = -z_critical               # Critical value negativo
print(f"Critical Value: {z_critical:.2f}")
```

```python
# Visualización
x = np.linspace(-4, 4, 1000)
y = st.norm.pdf(x, 0, 1)

plt.figure(figsize=(10, 6))
plt.plot(x, y, label="Normal Distribution", color="blue")
plt.axvline(z_critical, color="red", linestyle="--", label=f"Critical Value (+/-{z_critical:.2f})")
plt.axvline(z_critical_neg, color="red", linestyle="--")
plt.axvline(z_stat, color="green", linestyle="-", label=f"Z-Statistic ({z_stat:.2f})")

# Rellenar las regiones de rechazo
plt.fill_between(x, y, where=(x > z_critical) | (x < z_critical_neg), color="red", alpha=0.3, label="Rejection Regions")
plt.fill_between(x, y, where=(x <= z_critical) & (x >= z_critical_neg), color="blue", alpha=0.1, label="Acceptance Region")

plt.title("Z-Test Two-Tail Visualization")
plt.xlabel("Z-Score")
plt.ylabel("Probability Density")
plt.legend()
plt.grid()
plt.show()
```
